```python
import pandas as pd
import io
df=pd.read_csv(io.BytesIO(uploaded['kag_risk_factors_cervical_cancer.csv']))
print(df)
```

```python
df.isnull().any().any()
df = df.fillna(0)
```

```python
import numpy as np
df.replace(to_replace=np.nan,value=0)
```

```python
for feature in df.columns:
    df[feature].replace('?',np.nan,inplace=True )
    df[feature].fillna(value=0,inplace=True)
for feature in df.columns:
    df[feature].replace(0,df[feature].median(),inplace=True)
df.head()
```

```python
# Remove columns with all na values
all_na = df.columns[df.isna().all()]
df.drop(all_na, axis = 1, inplace = True)
```

```python
import numpy as np
import pandas as pd
import statsmodels.api as sm
import matplotlib.pyplot as plt
import seaborn as sns
sns.set()
from sklearn.cluster import KMeans
```

```python
x = df.iloc[:,0:2]
x
```

```python
kmeans = KMeans(2)
kmeans.fit(x)
```

```python
identified_clusters = kmeans.fit_predict(x)
identified_clusters
```

```python
df['Age'] = df['Age'].astype(str)
df['Number of sexual partners'] = df['Number of sexual partners'].astype(str)
```

```python
data_with_clusters = df.copy()
data_with_clusters['Clusters'] = identified_clusters 
plt.scatter(data_with_clusters['Age'],data_with_clusters['Number of sexual partners'],c=data_with_clusters['Clusters'],cmap='rainbow')
```

```python
wcss=[]
for i in range(1,7):
  kmeans = KMeans(i)
  kmeans.fit(x)
  wcss_iter = kmeans.inertia_
  wcss.append(wcss_iter)


number_clusters = range(1,7)
plt.plot(number_clusters,wcss)
plt.title('The Elbow title')
plt.xlabel('Number of clusters')
plt.ylabel('WCSS')
```
